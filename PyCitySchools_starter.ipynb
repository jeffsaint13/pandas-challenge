```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

# Pushing the data in to a DataFrame
school_dataset_df = pd.DataFrame(school_data_complete)
school_dataset_df.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
total_schools = (school_dataset_df["school_name"].value_counts())
total_schools.count()
```

```python
# Calculate the total number of students
total_students = (school_dataset_df["Student ID"].value_counts())
total_students.count()
```

```python
# Calculate the total budget

total_budget = (school_dataset_df["budget"].unique())
total_budget.sum()
```

```python
# Calculate the average math score

avg_math_score = (school_dataset_df["math_score"].mean())
avg_math_score
```

```python
# Calculate the average reading score
avg_reading_score = (school_dataset_df["reading_score"].mean())
avg_reading_score
```

```python
# Calculate the percentage of students with a passing math score (70 or greater)
percent_math_pass = school_dataset_df.loc[school_dataset_df["math_score"] >=70]["math_score"].count()
school_dataset_df["% Passing Math"] = percent_math_pass = percent_math_pass/total_students.count()*100
percent_math_pass
```

```python
# Calculate the percentage of students with a passing reading score (70 or greater)
percent_reading_pass = school_dataset_df.loc[school_dataset_df["reading_score"] >=70]["reading_score"].count()
school_dataset_df["% Passing Reading"] = percent_reading_pass = percent_reading_pass/total_students.count()*100
percent_reading_pass
```

```python
# Calculate the percentage of students who passed math and reading (% Overall Passing)

overall_pass = school_dataset_df[(school_dataset_df["math_score"] >=70) & (school_dataset_df["reading_score"] >=70)]["Student ID"].count()/total_students.count()*100
school_dataset_df["% Overall Passing"] = overall_pass
overall_pass
```

```python
# Create a dataframe to hold the above results
school_analysis_df = pd.DataFrame({'Total Schools': [total_schools.count()],
                                  'Total Students': [total_students.count()],
                                  'Total Budget': [total_budget.sum()],
                                  'Average Math Score': [avg_math_score],
                                  'Average Reading Score': [avg_reading_score],
                                  '% Passing Math': [percent_math_pass],
                                  '% Passing Reading': [percent_reading_pass],
                                  '% Passing Passing': [overall_pass]  
                                 })

# Use Map to format all the columns
school_analysis_df["Total Students"] = school_analysis_df["Total Students"].map("{:,.0f}".format)
school_analysis_df["Total Budget"] = school_analysis_df["Total Budget"].map("${:,.2f}".format)
school_analysis_df
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)

```python
# Calculate the Per Student Budget
school_dataset_df["Per Student Budget"] = school_dataset_df["budget"]/school_dataset_df["size"]
per_student_bud = ((school_dataset_df["Per Student Budget"]).mean())
per_student_bud
```

```python
# Complete grouping
school_grouping = school_data_complete.set_index("school_name").groupby(["school_name"])

# Count of students distant
total_stu = school_grouping["Student ID"].count()

# Calculate the percentage of students with a passing math score (70 or greater)
percent_math_pass = school_data_complete[school_data_complete["math_score"] >=70].groupby("school_name")["Student ID"].count()/total_stu*100

# Calculate the percentage of students with a passing reading score (70 or greater)
percent_reading_pass = school_data_complete[school_data_complete["reading_score"] >=70].groupby("school_name")["Student ID"].count()/total_stu*100

# Calculate the percentage of students who passed math and reading (% Overall Passing)
percent_overall_pass = school_data_complete[(school_dataset_df["math_score"] >=70) & (school_dataset_df["reading_score"] >=70)].groupby("school_name")["Student ID"].count()/total_stu*100



percent_grouping = pd.DataFrame({
                                "% Passing Math": percent_math_pass,
                                "% Passing Reading": percent_reading_pass,
                                "% Overall Passing": percent_overall_pass
                                })

percent_grouping
```

```python
school_data_df = school_dataset_df[["school_name", "type","size","budget","Per Student Budget", "math_score", "reading_score"]]

# Grouping by school type
school_grouping = school_data_df.groupby(["school_name", "type"]).mean()


# Renaming column names
school_summary = school_grouping.rename(columns= {"budget": "Total School Budget",
                                                       "type": "School Type",
                                                       "math_score": "Average Math Score",
                                                       "reading_score": "Average Reading Score",
                                                       "size": "Total Students"
                                                        })

# Use Map to format all the columns
# school_summary["Total Students"] = school_summary["Total Students"].map("{:,.0f}".format)
school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
# school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:.2f}".format)
school_summary
```

```python
# Completing my merge of the overall school percentages with summary

overall_sch_summary = pd.merge(school_summary, percent_grouping, on="school_name")
overall_sch_summary
```

## Math Scores by Grade

* Create a table that lists the average Reading Score for students of each grade level (9th, 10th, 11th, 12th) at each school.

  * Create a pandas series for each grade. Hint: use a conditional statement.
  
  * Group each series by school
  
  * Combine the series into a dataframe
  
  * Optional: give the displayed data cleaner formatting

```python
math_grouping = school_data_complete.groupby(["school_name","grade"]).agg({"math_score":["mean"]}).unstack()

# Reordering Columns
math_grouping = math_grouping[[('math_score', 'mean',  '9th'),('math_score', 'mean',  '10th'),('math_score', 'mean',  '11th'),('math_score', 'mean',  '12th')]]


# Use Map to format all the columns
math_grouping[('math_score', 'mean',  '9th')] = math_grouping[('math_score', 'mean',  '9th')].map("{:.2f}".format)
math_grouping[('math_score', 'mean',  '10th')] = math_grouping[('math_score', 'mean',  '10th')].map("{:.2f}".format)
math_grouping[('math_score', 'mean',  '11th')] = math_grouping[('math_score', 'mean',  '11th')].map("{:.2f}".format)
math_grouping[('math_score', 'mean',  '12th')] = math_grouping[('math_score', 'mean',  '12th')].map("{:.2f}".format)

# Pring out
math_grouping
```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python

# Create the bins in which data will be held
bins = [0.0,585.0,630.0,645.0,680.0]

# Create the names for the five bins
group_names = ["<$585","$585-630","$630-645","$645-680"]

overall_sch_summary["Spending Ranges (Per Student)"] = pd.cut(overall_sch_summary["Per Student Budget"],bins, labels=group_names, include_lowest=True)

# Creating a group based off of the bins
school_spending = overall_sch_summary.groupby("Spending Ranges (Per Student)").mean()


# Use Map to format all the columns
school_spending["Average Math Score"] = school_spending["Average Math Score"].map("{:.2f}".format)
school_spending["Average Reading Score"] = school_spending["Average Reading Score"].map("{:.2f}".format)
school_spending["% Passing Math"] = school_spending["% Passing Math"].map("{:.2f}".format)
school_spending["% Passing Reading"] = school_spending["% Passing Reading"].map("{:.2f}".format)
school_spending["% Overall Passing"] = school_spending["% Overall Passing"].map("{:.2f}".format)
school_spending["Per Student Budget"] = school_spending["Per Student Budget"].map("${:.2f}".format)
school_spending["Total Students"] = school_spending["Total Students"].map("{:,.0f}".format)


school_spending
```
